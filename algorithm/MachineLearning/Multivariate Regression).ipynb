```python
from sklearn.preprocessing import LabelEncoder

encoder1 = LabelEncoder()
DF['sex'] = encoder1.fit_transform(DF.sex)
```

```python
encoder2 = LabelEncoder()
DF['smoker'] = encoder2.fit_transform(DF.smoker)
```

```python
encoder3 = LabelEncoder()
DF['region'] = encoder3.fit_transform(DF.region)
```

```python
DF.info()
```

```python
DF.head()
```

```python
from sklearn.model_selection import train_test_split

X = DF[['age', 'sex']]
y = DF['expenses']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2045 ) 

print('Train Data :', X_train.shape, y_train.shape)
print('Test Data :', X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

RA = LinearRegression()
RA.fit(X_train, y_train)
```

```python
y_hat = RA.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse1 = mean_squared_error(y_test,y_hat)
np.sqrt(mse1)
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/Insurance.csv'
DF = pd.read_csv(url)
DF.info()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(DF,
                                       test_size = 0.3,
                                       random_state = 2045)
train_set.shape, test_set.shape
```

```python
train_set.info()
```

```python
import statsmodels.formula.api as smf

Model_1 = smf.ols(formula = 'expenses ~ age + sex', data = train_set).fit()
```

```python
y_hat_1 = Model_1.predict(test_set[['age','sex']])
```

```python
mse2 = mean_squared_error(test_set.expenses, y_hat_1)
np.sqrt(mse2)
```

```python
print('sklearn     :', np.sqrt(mse1))
print('statsmodels :', np.sqrt(mse2))
```
