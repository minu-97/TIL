```python
from apscheduler.schedulers.blocking import BlockingScheduler #단일스케쥴러
from apscheduler.schedulers.background import BackgroundScheduler #다중스케쥴러
import time
from urllib.request import urlopen #python 내장 라이브러리
import urllib
import bs4
import numpy as np
def job1():
    print('크롤링 수행시간 :', time.strftime('%H시 %M분 %S초',time.localtime(time.time())))

def job2():
    #IT_디지털데일리
    #1. html 가져오기 
    # 메인 페이지
    url_list2 = []
    p = 'https://media.naver.com/press/'
    uid_table = ['032','005','020','021','081','022','023','025','028','001']
    sid_table = [100, 101, 102, 104, 105]
    for uid_t  in uid_table:
        uid = uid_t+'?'
        for i in sid_table:
            sid = 'sid='+ f"{i}"
            url = p+uid+sid
            html = urllib.request.urlopen(url)
            bs = bs4.BeautifulSoup(html, 'html.parser')
            #print(bs)
            # 언론사별 총 링크
            #주요뉴스
            list = bs.findAll('a',{'class':'press_edit_news_link _es_pc_link'})
            for n, txt in enumerate(list):
                url_list2.append(txt['href'])
    print(url_list2)
    return url_list2
    
def main():
    #scheduler1 = BlockingScheduler()
    scheduler2 = BackgroundScheduler()
    
    scheduler2.add_job(job1, 'cron', hour ='17', minute = '12', second ='10' ,id = 
                       'test1')
    scheduler2.add_job(job2, 'cron', hour ='17', minute = '12',second ='10' ,id = 
                       'test2')
    scheduler2.start()
    
    
print('operation start')


if __name__ == "__main__":
    main()
```
